```python
import pandas as pd
import numpy as np
import os
```

```python
dir += '\Datasets'
dir
```

```python
# Read and store content
# of an excel file 
read_file = pd.read_excel(dir + '\Births.xlsx')
  
# Write the dataframe object
# into csv file
read_file.to_csv(dir+"\Births.csv", 
                  index = None,
                  header=True)
    
# read csv file and convert 
# into a dataframe object
df_births = pd.DataFrame(pd.read_csv(dir+"\Births.csv"))
  
# show the dataframe
df_births
```

```python
for index, row in df_births.iterrows():
    for year in df_births.columns:
        if df_births[year][index] == ':': df_births[year][index]  = None
df_births
```
