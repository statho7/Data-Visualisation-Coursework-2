## Preprocess the data for all the dataframes

```python
def Preprocess(df):
    '''
    This function preprocesses the given dataframe to include only the EU28 countries. The rest of the countries are dropped by the dataframe and the columns European Union - 28 countries (2013-2020) & Germany (until 1990 former territory of the FRG) are changing names to European Union and Germany respectively.
    
    Args:
        df: The dataframe we want to preprocess.

    Returns: The modified dataframe.
    '''

    df['Country'][1] = 'European Union'
    df['Country'][9] = 'Germany'

    EU28 = ['Austria','Belgium','Bulgaria','Croatia','Cyprus','Czechia','Denmark','Estonia',
    'Finland','France','Germany','Greece','Hungary','Ireland','Italy','Latvia','Lithuania',
    'Luxembourg','Malta','Netherlands','Poland','Portugal','Romania','Slovakia','Slovenia',
    'Spain','Sweden','United Kingdom']

    if '2008' in df.columns:
        df = df.drop(['2008'], axis = 1)

    # Empty array to append here the indexes of the countries that do not belong to the EU28 and we will drop from the dataframe
    indexes = []
    # Iterate through the dataframe
    for index, row in df.iterrows():
        # Check if country is in EU28 and index different to 1 as we want to keep the data for European Union - 28 countries (2013-2020)
        if df['Country'][index] not in EU28 and index != 1:
            # If country is not in EU28 we append its index to indexes array
            indexes.append(index)
        else:
            # Iterate through columns (years)
            for year in df.columns[1:]:
                # If a cell for a country that belongs to EU28 has as value the ':' we convert it to None
                # If not we are converting the string value to a float one
                if df[year][index] == ':': 
                    df[year][index]  = None
                else:
                    df[year][index] = float(df[year][index])

    # Drop countries that do not belong in EU28
    df = df.drop(df.index[[indexes]])
    return df
```

### Call the Preprocess function

```python
df_births = Preprocess(df_births)
df_deaths = Preprocess(df_deaths)
df_Pop_0_14 = Preprocess(df_Pop_0_14)
df_Pop_15_24 = Preprocess(df_Pop_15_24)
df_Pop_25_49 = Preprocess(df_Pop_25_49)
df_Pop_50_64 = Preprocess(df_Pop_50_64)
df_Pop_65_79 = Preprocess(df_Pop_65_79)
df_Pop_80 = Preprocess(df_Pop_80)
df_population = Preprocess(df_population)
```
